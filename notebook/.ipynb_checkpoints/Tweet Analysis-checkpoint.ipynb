### IMPORT LIBRARIES

```python
import pandas as pd
import re 
import tweepy 
from tweepy import OAuthHandler 
from textblob import TextBlob 
from wordcloud import WordCloud
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, '../code/')
from TwitterAPI import TwitterClient
```

### HILLARY CLINTON vs DONALD TRUMP DATASET

```python
# define function to analyze twitter dataset
def main_from_df(author): 
    
    api = TwitterClient()
    
    tweets_list = api.get_tweets_from_df(author)
    
    tweets_sentiment = api.analyze_tweets(tweets_list) 
    
    api.print_tweet_sentiment(tweets_sentiment)
    
    return tweets_list
```

```python
# Trump tweets analysis
trump_tweets = main_from_df('realDonaldTrump')
```

```python
# Clinton tweets analysis
clinton_tweets = main_from_df('HillaryClinton')
```

### TEST ON CUSTOM TWEETS

```python
# function to test the algorithm on new tweets
def main_from_user(tweets_list): 
    
    api = TwitterClient()
    
    tweets_sentiment = api.analyze_tweets(tweets_list) 
    
    api.print_tweet_sentiment(tweets_sentiment)
    
    return
```

```python
# calling main function
my_tweets = ['i hate you all', 'i love you all', 'i don t care']
    
main_from_user(my_tweets)
```

### WORDCLOUD TRUMP

```python
text= ' '.join(trump_tweets)
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
